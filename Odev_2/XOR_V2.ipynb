```python
import numpy as np

X = np.array([[0,0],[0,1],[1,0],[1,1]])#Girdi değerleri
y = np.array([[0],[1],[1],[0]])# çıktı değerleri

# Tek katmanlı NN (perceptron yapısı)
np.random.seed(42)
Weight = np.random.randn(2, 1) * 0.1 #Ağırlıkların iilk değerlerinin oluşturulması
bias = np.zeros((1, 1)) # bias değerlerinin oluşturulması
lr = 0.5 # ogrenme hızı, katsayısı

#Burada perceptron içerisindeki aktivasyon fonksiyonu olarak sigmoid seçtim
def sigmoid(x):
    return 1 / (1 + np.exp(-np.clip(x, -500, 500))) # Taşma hatası alınmasın diye sınırlama yaptım

#Eğitim döngüsü
for _ in range(2000):
    out = sigmoid(X @ Weight + bias)#Burada 4x2 matrisi olan X değerleri ile 2x1 matrisi olan W ağırlıklarının matris çarpımı var + bias
    Weight -= lr * (X.T @ (out - y)) / len(X) # ağırlıkların güncellenmesi
    bias -= lr * np.mean(out - y)

#Tahminler
preds = (sigmoid(X @ Weight + bias) > 0.5).astype(int)

for i in range(4):
    print(f"{X[i]} => beklenen: {y[i][0]}, tahmin: {preds[i][0]}")
print(f"Doğruluk: {np.mean(preds == y)*100:.1f}%")
```

```python
# Çok katmanlı

#Aynı şekilde ağırlık ve bias değerlerinin oluşturulması tek fark ağırlık öşçeklendirmeyi katmanlardaki nöron sayısı ile yaptım
np.random.seed(42)

#giriş katmanı ağırlık ve bias değeri
W1 = np.random.randn(2, 4) * np.sqrt(2/2) # gizlii katmanda 4 nöron olduğu için fark var
b1 = np.zeros((1, 4)) # ilk bağlantı kısmındaki bias değerleri

#Çıkış katmanı ağırlık ve bias değeri
W2 = np.random.randn(4, 1) * np.sqrt(2/4)
b2 = np.zeros((1, 1))
lr = 0.5

def tanh_d(x): return 1 - np.tanh(x)**2

def sigmoid_d(a): return a * (1 - a)

#Eğiitim
for _ in range(10000):
    #ilk olarak ileri yayılım yaptım
    z1 = X @ W1 + b1
    a1 = np.tanh(z1) # Gizli katman çıktısı
    z2 = a1 @ W2 + b2
    a2 = sigmoid(z2) # output çıktısı

    # geri yayılım ( türev, zincir kuralı )
    dz2 = (a2 - y) * sigmoid_d(a2) # burada sigmoidin türevi ile tahmin ile gerçek farkını çarpıyoruz bu şekilde son katmanın toplamının hatasını öğreniyoruz
    W2 -= lr * (a1.T @ dz2) / len(X) # Gradient hesap edilip ağırlıklar değiştirilir
    b2 -= lr * np.mean(dz2)

    dz1 = (dz2 @ W2.T) * tanh_d(z1) # burada tanh_d kullanma sebebi gizli katmanın aktivasyon donk olarak tanh kullanılmasıdır
    W1 -= lr * (X.T @ dz1) / len(X)
    b1 -= lr * np.mean(dz1, axis=0)

#tahminler
preds = (a2 > 0.5).astype(int)

#Çıktı
for i in range(4):
    print(f"{X[i]} => beklenen: {y[i][0]}, tahmin: {preds[i][0]}")
print(f"Doğruluk: {np.mean(preds == y)*100:.1f}%")
```
