```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Input

# XOR verisi
inputs  = np.array([[0,0],[0,1],[1,0],[1,1]])
outputs = np.array([[0],[1],[1],[0]])

# Model
model = Sequential()
model.add(Input(shape=(2,)))
model.add(Dense(4, activation='tanh'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Eğitim
history = model.fit(inputs, outputs, epochs=10000, verbose=0)

# Metrikler
loss, acc = model.evaluate(inputs, outputs, verbose=0)
tahminler = model.predict(inputs, verbose=0).flatten()

print(f"Son Kayıp  : {loss:.4f}")
print()
print("Giriş  | Gerçek | Tahmin")
print("-" * 30)

for i, (x, y) in enumerate(zip(inputs, outputs.flatten())):
    t = tahminler[i]
    print(f"{x}  |   {y}    |  {t:.3f}")
```
